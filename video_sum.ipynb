```python
import subprocess
import json
import os

video_url = "https://www.youtube.com/watch?v=bhPMvyNuQq8"
video_id = video_url.split("v=")[1]
caption_file = f"{video_id}.en.json3"

# Try downloading captions
result = subprocess.run([
    "yt-dlp",
    "--skip-download",
    "--write-auto-sub",  # auto-generated captions
    "--sub-lang", "en",
    "--sub-format", "json3",
    f"--output={video_id}.%(ext)s",
    video_url
], capture_output=True, text=True)

if result.returncode != 0:
    print("yt-dlp failed:")
    print(result.stderr)
else:
    if not os.path.exists(caption_file):
        raise FileNotFoundError(f"No captions found for video: {video_url}")

    with open(caption_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    full_text = " ".join(
        seg["utf8"]
        for event in data.get("events", [])
        if "segs" in event
        for seg in event["segs"]
    )
    print(full_text)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain


llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    temperature=0.3,
    google_api_key="AIzaSyAajBOEkyYfhW87a7Wn9e05KDKckAiBeD8"
)

prompt = PromptTemplate(
    input_variables=["transcript"],
    template="""
You are an expert summarizer.
Summarize the following YouTube transcript in 5 concise bullet points:

Transcript:
{transcript}
"""
)

chain = LLMChain(llm=llm, prompt=prompt)
summary = chain.run({"transcript": full_text})

print("\n--- SUMMARY ---")
print(summary)
```
